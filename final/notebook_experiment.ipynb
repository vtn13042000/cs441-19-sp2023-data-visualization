```python
# function to round the number
def round( n ):
  
    # Smaller multiple
    a = (n // 10) * 10
      
    # Larger multiple
    b = a + 10
      
    # Return of closest of two
    return (b if n - a > b - n else a)
  
# driver code
n = max(df.Close.max())
print(round(n))
```
